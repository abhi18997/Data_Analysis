```python
#loading the packages required for the analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Read the csv file to understand the data**

```python
df_tmdb = pd.read_csv('tmdb-movies.csv')
df_tmdb
```

**-DATA WRANGLING**

**Data cleaning process**:

**1)Check for the null values present in the dataset**

**2)Check for duplicate value present in the dataset**

**3)Removing all the unwanted columns which are not useful for cleaning from dataset**

**4)Change the datetime format of dataset**

**5)Replace all the null or zero values**

**6)Which movie has minimum profit?**

**7)Which movie has maximum profit?**

**


```python
#To check null values and column from data set
df_tmdb.info()
```

```python
#To fill for all missing values from data set
df_tmdb.fillna(df_tmdb.mean(),inplace = True)
df_tmdb.info()
```

```python
# check for duplicates in the data
sum(df_tmdb.duplicated())
```

```python
# drop duplicates
df_tmdb.drop_duplicates(inplace=True)
```

```python
# confirm correction by rechecking for duplicates in the data
sum(df_tmdb.duplicated())
```

```python
#Removing unwanted columns from the dataset
df_tmdb.drop(['id','imdb_id','homepage','keywords','overview','production_companies','tagline','budget_adj','revenue_adj','vote_count','vote_average'], axis=1, inplace=True)
df_tmdb
```

```python
#replace zeros with null values in the budget and revenue column
df_tmdb['budget']= df_tmdb['budget'].replace(0,np.NaN)
df_tmdb['revenue']= df_tmdb['revenue'].replace(0,np.NaN)
df_tmdb['cast']= df_tmdb['cast'].replace(0,np.NaN)
df_tmdb['runtime']= df_tmdb['runtime'].replace(0,np.NaN)
df_tmdb.info()
```

```python
#changing the release date column into standard date format
df_tmdb.release_date = pd.to_datetime(df_tmdb['release_date'])
df_tmdb.head()
```

```python
#checking the data type of dataset
df_tmdb.dtypes
```

**As the data type of dataset is correct, so there is no need to change any data type from dataset**

**Research Question 1) What is the proift of the movies from the dataset over the years?**

```python
#Calculating profit of each movies from dataset
df_tmdb.insert(3,'profit',df_tmdb['revenue']-df_tmdb['budget'])
df_tmdb.head()
```

**Research question 2) Which is the worst year amongst all the years to release movie in theatre or make profit?**

```python
#Worst year to release movie in theatre
df_tmdb[df_tmdb['release_year'] == df_tmdb['release_year'].min()]
```

From above analysis we can say that 1960 is the worst year to releast the movies in theatres or make profit 

**Research Question 3) Year of Release Vs the Profitability of the movies**

```python
profit_year = df_tmdb.groupby('release_year')['profit'].sum()
plt.figure(figsize = (12,6))

plt.xlabel('Year')
plt.ylabel('Profit')
plt.title('Total profit earned by movies over years')
plt.plot(profit_year)
plt.show()
```

Now from above graph we can conclude that the 2015 year was the year when movies made the highest profit after releasing in the theatre.

**Research Question 4) What is the average run time of all movies from dataset?**

```python
def avg_fun(column):
    return df_tmdb[column].mean()
```

```python
avg_fun('runtime')
```

```python
plt.figure(figsize = (9,5), dpi= 100)

plt.xlabel('Runtime', fontsize = 15)
plt.ylabel('Number of movies', fontsize = 15)
plt.title('Run time of all movies', fontsize = 15)
plt.hist(df_tmdb['runtime'], bins = 20);
```

From above graph and analysis we can conclude that the average run time all the movies from the dataset is in between 80 to 150 minutes.

**Research Question 5) What are the movies genres present in the dataset? and which are the most popular genre year to year to make movies?**

```python
genres = df_tmdb['genres'].unique()
print(genres)
```

```python
def separate_count(column):
    split_data_tmdb = pd.Series(df_tmdb[column].str.cat(sep = '|').split('|'))
    
    count_data_tmdb = split_data_tmdb.value_counts(ascending=False)
    return count_data_tmdb
```

```python
separate_count("genres").plot(kind="pie",figsize=(9,9))
plt.title('Popular Genres')
plt.ylabel('');
```

From above analysis we can conclude that drama is most popularly used genre to make movies.Because of which most profitable movies are made of drama genre and then comedy.

**Research Question 8) What us relaton between budget and revenue?**

```python
budget_revenue = df_tmdb.groupby('budget')['revenue'].sum()
plt.figure(figsize = (12,6))

plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs Revenue of movies')
plt.plot(budget_revenue)
plt.show()
```
